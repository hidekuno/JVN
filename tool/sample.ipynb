#### 2次元アフィン変換

$$ \left(
\begin{array}{c}
x' \\
y' \\
1
\end{array} 
\right) 
=
\left( 
\begin{array}{ccc} 
m_{11} & m_{21} & d_{x} \\
m_{12} & m_{22} & d_{y} \\
0 & 0 & 1
\end{array} 
\right)
\left( 
\begin{array}{c}
x \\
y \\
1
\end{array} 
\right) $$

#### 回転移動 $$
\begin{pmatrix}
  x' \\ y'
\end{pmatrix}
=
\begin{pmatrix}
   \cos\theta & \sin\theta \\
  -\sin\theta & \cos\theta
\end{pmatrix}
\begin{pmatrix}
  x \\ y
\end{pmatrix}
$$
#### 平行移動 $$
\begin{pmatrix}
  x' \\ y'
\end{pmatrix}
=
\begin{pmatrix}
  x \\ y
\end{pmatrix}
+
\begin{pmatrix}
  t_{x} \\ t_{y}
\end{pmatrix}
$$
#### 拡大・縮小 $$
\begin{pmatrix}
  x' \\ y'
\end{pmatrix}
=
\begin{pmatrix}
  s_{x} & 0 \\ 0 & s_{y}
\end{pmatrix}
\begin{pmatrix}
  x \\ y
\end{pmatrix}
$$
